### Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = df ,x = 'airline',y='price')
plt.xticks(rotation=45,ha = 'right')
plt.xlabel("Airline")
plt.ylabel("Ticket Price")
plt.title("Price Variation Per Airline")
plt.show()
```

### Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = df ,x = 'airline',y='price')
plt.xticks(rotation=45,ha = 'right')
plt.xlabel("Airline")
plt.ylabel("Ticket Price")
plt.title("Price Variation Per Airline")
plt.show()
```

**Yes, there are outliers in the dataset, and the boxplot provided clearly shows them.**

1.In a boxplot, outliers are typically the individual points that are plotted outside the "whiskers" (the lines extending from the box).

2.For airlines like SpiceJet, AirAsia, GO_FIRST, and Indigo, you can see several data points above the top whisker  these are outliers indicating unusually high ticket prices compared to the rest of the data.

3.Even for Vistara and Air India, which already have a wide range of ticket prices (larger boxes and whiskers), there are still some outliers far above the usual range.

#### How outliers may impact analysis:

1.Mean ticket price will be skewed upwards because outliers increase the average.

2.Regression models or predictive models may become less accurate if outliers are not handled because the model might overfit to these extreme values.

3.Interpretations about the general affordability of airlines could be misleading if you don't account for these unusually high prices.
