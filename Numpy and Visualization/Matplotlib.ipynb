### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.


**Matplotlib** is a popular data visualization library in Python. It provides tools for creating static, interactive, and animated plots and graphs. Communicate findings through graphs

Create customizable and high-quality plots

Export visuals in multiple formats (e.g., PNG, PDF, SVG)

#### Five common plots using matplotlib.pyplot:

Line Plot – to show trends over time or continuous data

Scatter Plot – to explore relationships between two numeric variables

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data  plot a scatter plot.   
import numpy as np 
np.random.seed(3)  
x = 3 + np.random.normal(0, 2, 50)  
y = 3 + np.random.normal(0, 2, len(x))   
Note: Also add title, xlabel, and ylabel to the plot. 

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.xlabel("This is Value of x plotted on X-Axis")
plt.ylabel("This is Value of y plotted on Y-Axis")
plt.title("Scatter Plot")
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.  
Use the following data:  
import numpy as np  
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])  
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])  
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])  
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])  

Ans)The **subplot() function** in Matplotlib is used to create multiple plots (subplots) within a single figure. It allows you to arrange multiple plots in rows and columns, creating a grid-like layout. This function is useful when you want to compare multiple plots side by side or visualize different aspects of the data in one figure.

matplotlib.pyplot.subplot(nrows, ncols, index)


Parameters:
nrows: The number of rows of subplots in the grid.
ncols: The number of columns of subplots in the grid.
index: The position of the subplot within the grid (starting from 1 and incrementing from left to right, top to bottom).

```python
import numpy as np
import matplotlib.pyplot as plt

#For line 1: 
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
#For line 2: 
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
#For line 3: 
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
#For line 4: 
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.  
import numpy as np  
company = np.array(["Apple", "Microsoft", "Google", "AMD"])  
profit = np.array([3000, 8000, 1000, 10000])   

Ans)A **bar plot**, also known as a bar chart or column chart, is a type of graph that represents categorical data using rectangular bars. Each bar's length or height is proportional to the value it represents, and the bars are typically arranged along the x-axis for easy comparison between categories.

#### Bar plots are commonly used for the following purposes:

1.Visual Comparison: Bar plots are effective in visually comparing different categories and their corresponding values. They allow us to identify trends, patterns, and differences between different groups easily.

2.Categorical Data Representation: Bar plots are suitable for displaying discrete or categorical data. They can show the frequency, count, or any other measure associated with each category.

3.Visualizing Distributions: Bar plots can be used to display the distribution of data across categories. They provide a clear representation of the distribution of values for each category.

4.Presentation of Rankings: Bar plots can be used to present rankings or ordering of categories based on their values. The taller (or longer) bars indicate higher values or ranks.

5.Comparison of Data Over Time: Bar plots can be used to compare data across different time periods or intervals when multiple sets of data are available.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Bar Graph")
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.  
box1 = np.random.normal(100, 10, 200)  
box2 = np.random.normal(90, 20, 200)  

Ans) A box plot, also known as a box-and-whisker plot, is a type of statistical visualization that displays the distribution of a dataset. It provides a concise summary of the data's central tendency, spread, and the presence of outliers. The box plot is particularly useful for identifying the range of values, the presence of outliers, and the symmetry or skewness of the data.

#### The main components of a box plot are as follows:

**Box:** The box in the plot represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3). The central line inside the box represents the median (Q2), which is the middle value when the data is sorted.  
**Whiskers:** The whiskers extend from the edges of the box to the minimum and maximum values within a certain range, usually determined by a factor called the "whisker length." These whiskers provide information about the spread of the data outside the IQR.  
**Outliers:** Outliers are data points that lie significantly beyond the whiskers, indicating extreme values in the dataset. 

#### Box plots are used for the following purposes:

1.Data Distribution: Box plots provide a visual representation of the data's distribution, revealing whether it is symmetrical, skewed to the left, or skewed to the right. The central line (median) indicates the data's central tendency, while the box and whiskers provide information about the data's spread.

2.Comparison of Groups: Box plots are useful for comparing the distributions of different groups or categories in a dataset. By plotting multiple box plots side by side, it becomes easy to identify differences in the central tendency and spread of different groups.

3.Outlier Detection: Box plots help identify potential outliers in the data by displaying data points beyond the whiskers. Outliers can be valuable for understanding extreme values or errors in the data.

4.Data Summary: Box plots provide a concise summary of the data's characteristics without displaying each data point, making them useful when dealing with large datasets.

5.Identifying Skewness: The shape of the box plot can indicate whether the data is symmetrically distributed or skewed to one side, providing insights into the data's underlying distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create box plot
data = [box1, box2]
plt.boxplot(data, labels=['Box 1', 'Box 2'])

# Add title and labels
plt.title('Box Plot Example')
plt.ylabel('Values')

# Show plot
plt.show()
```
